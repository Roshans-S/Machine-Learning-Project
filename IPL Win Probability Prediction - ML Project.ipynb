## Merging Deliveries and Match_df

```python
delivery_df = match_df.merge(deliveries,on='match_id')
delivery_df.head()
```

```python
delivery_df.shape
```

```python
# Filtering on the basis of 2nd innning
delivery_df_2inng = delivery_df[delivery_df['inning']==2]
delivery_df_2inng.head()
```

```python
# Calculating the current score

import warnings
warnings.filterwarnings("ignore")

delivery_df_2inng['Current_Score'] = delivery_df_2inng.groupby('match_id')[['total_runs_y']].cumsum()
delivery_df_2inng.head()
```

```python
# Calculating the runs left
delivery_df_2inng['runs_left'] = delivery_df_2inng['total_runs_x'] - delivery_df_2inng['Current_Score']
delivery_df_2inng.head()
```

```python
# Calculating balls left 

delivery_df_2inng['Balls left'] = 126 -((delivery_df_2inng['over']*6)+delivery_df_2inng['ball'])
delivery_df_2inng.head()
```

```python
# Checking for nan values in player dismissed column
delivery_df_2inng['player_dismissed'].isnull().sum()
```

```python
# Filling the nan values with 0 in player dismissed column
delivery_df_2inng['player_dismissed'] = delivery_df_2inng['player_dismissed'].fillna(0)
```

```python
delivery_df_2inng['player_dismissed'].isnull().sum()
```

```python
# Converting Player _dismissed column to boolean , 1 - Dismissed , 0- not dismissed.
delivery_df_2inng['player_dismissed'] = delivery_df_2inng['player_dismissed'].apply(lambda x:x if x == 0 else 1)
```

```python
delivery_df_2inng.head()
```

```python
# Calculating wickets left
wicket = delivery_df_2inng.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
wicket
```

```python
delivery_df_2inng['wicket'] = 10-wicket
```

```python
delivery_df_2inng.tail()
```

```python
# Calculating current run rate
delivery_df_2inng['Current Run_rate'] = (delivery_df_2inng['Current_Score']*6)/(120-delivery_df_2inng['Balls left'])
```

```python
delivery_df_2inng.head()
```

```python
# Calculating Required run rate
delivery_df_2inng['Req Run_Rate'] = (delivery_df_2inng['runs_left']*6)/(delivery_df_2inng['Balls left'])
```

```python
delivery_df_2inng.head()
```

```python
# Result

def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df_2inng['match result'] = delivery_df_2inng.apply(result, axis=1)
```

```python
delivery_df_2inng.head()
```

```python
delivery_df_2inng.columns
```

```python
final_df = delivery_df_2inng[['batting_team','bowling_team','city','runs_left','Balls left','wicket','total_runs_x','Current Run_rate','Req Run_Rate','match result']]
```

## Defining X and Y

```python
y = final_df[['match result']]
x = final_df.drop(columns=['match result'])
```

## Split the data into train and test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
x_train.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ("trf", OneHotEncoder(sparse=False, drop = 'first'),['batting_team', 'bowling_team', 'city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
ppline = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

```python
ppline.fit(x_train, y_train)
```

```python
y_predict = ppline.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```

### Accuracy from Logistic Regression model is 80.4%

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ppline2 = Pipeline(steps=[
    ('step1', trf),
    ('step2', RandomForestClassifier())
])
```

```python
ppline2.fit(x_train, y_train)
```

```python
y_predict1 = ppline2.predict(x_test)
```

```python
y_predict1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict1)
```

### The accuracy from Random Forest model is higher than the accuracy , got from Logistic Regression
